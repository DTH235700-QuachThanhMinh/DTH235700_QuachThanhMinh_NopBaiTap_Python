```python
#Câu1:
# Câu 1
def LuuFile(path, data):
    with open(path, 'a', encoding='utf-8') as file:
        file.write(data + "\n")

def DocFile(path):
    arrProduct = []
    with open(path, 'r', encoding='utf-8') as file:
        for line in file:
            data = line.strip()
            arr = data.split(';')
            arrProduct.append(arr)
    return arrProduct

# Nhập thông tin sản phẩm
masp = input("Nhập mã SP: ")
tensp = input("Nhập tên SP: ")
dongia = float(input("Nhập giá: "))

line = masp + ";" + tensp + ";" + str(dongia)

# Lưu vào file
LuuFile("database.txt", line)

# Đọc file
dssp = DocFile("database.txt")

# Xuất sản phẩm
def XuatSanPham(dssp):
    for row in dssp:
        for element in row:
            print(element, end="\t")
        print()
    print()

XuatSanPham(dssp)

# Sắp xếp theo đơn giá
def SortSp(dssp):
    for i in range(len(dssp)):
        for j in range(i + 1, len(dssp)):
            if float(dssp[i][2]) > float(dssp[j][2]):
                dssp[i], dssp[j] = dssp[j], dssp[i]
SortSp(dssp)

print("Sản phẩm sau khi sắp xếp giá:")
XuatSanPham(dssp)
```

```python
#Câu9:
import os

FILE_NAME = "sanpham.txt"

class DanhMuc:
    def __init__(self, ma, ten):
        self.ma = ma
        self.ten = ten

class SanPham:
    def __init__(self, ma, ten, dongia, madm):
        self.ma = ma
        self.ten = ten
        self.dongia = dongia
        self.madm = madm

danhmucs = []
sanphams = []

# -----------------------
# LƯU & ĐỌC FILE TEXT
# -----------------------

def save_to_file():
    with open(FILE_NAME, "w", encoding="utf-8") as f:
        for dm in danhmucs:
            f.write(f"DM|{dm.ma}|{dm.ten}\n")
        for sp in sanphams:
            f.write(f"SP|{sp.ma}|{sp.ten}|{sp.dongia}|{sp.madm}\n")
    print("Đã lưu file!")

def load_from_file():
    if not os.path.exists(FILE_NAME):
        return

    with open(FILE_NAME, "r", encoding="utf-8") as f:
        for line in f:
            data = line.strip().split("|")
            if data[0] == "DM":
                danhmucs.append(DanhMuc(data[1], data[2]))
            elif data[0] == "SP":
                sanphams.append(SanPham(data[1], data[2], float(data[3]), data[4]))

def show_all():
    print("\n===== DANH MỤC =====")
    for dm in danhmucs:
        print(dm.ma, dm.ten)

    print("\n===== SẢN PHẨM =====")
    for sp in sanphams:
        print(sp.ma, sp.ten, sp.dongia, "Danh mục:", sp.madm)


# -----------------------
# CHỨC NĂNG CRUD
# -----------------------

def add_danhmuc():
    ma = input("Mã DM: ")
    ten = input("Tên DM: ")
    danhmucs.append(DanhMuc(ma, ten))
    print("Thêm danh mục thành công!")

def add_sanpham():
    ma = input("Mã SP: ")
    ten = input("Tên SP: ")
    dongia = float(input("Đơn giá: "))
    madm = input("Mã danh mục: ")
    sanphams.append(SanPham(ma, ten, dongia, madm))
    print("Thêm sản phẩm thành công!")

def sort_sanpham():
    sanphams.sort(key=lambda x: x.ten)
    print("Đã sắp xếp theo tên!")

def search_sp():
    keyword = input("Nhập tên SP cần tìm: ")
    for sp in sanphams:
        if keyword.lower() in sp.ten.lower():
            print(sp.ma, sp.ten, sp.dongia, sp.madm)

# -----------------------
# MENU
# -----------------------

load_from_file()

while True:
    print("\n=== MENU QUẢN LÝ SẢN PHẨM ===")
    print("1. Thêm danh mục")
    print("2. Thêm sản phẩm")
    print("3. Xem tất cả")
    print("4. Tìm kiếm sản phẩm")
    print("5. Sắp xếp sản phẩm theo tên")
    print("6. Lưu file")
    print("0. Thoát")

    c = input("Chọn: ")

    if c == "1": add_danhmuc()
    elif c == "2": add_sanpham()
    elif c == "3": show_all()
    elif c == "4": search_sp()
    elif c == "5": sort_sanpham()
    elif c == "6": save_to_file()
    elif c == "0":
        save_to_file()
        break
```

```python
#Câu10:
import json
import os

FILE_JSON = "sinhvien.json"

class Lop:
    def __init__(self, ma, ten):
        self.ma = ma
        self.ten = ten

class SinhVien:
    def __init__(self, ma, ten, namsinh, malop):
        self.ma = ma
        self.ten = ten
        self.namsinh = namsinh
        self.malop = malop

lops = []
sinhviens = []

# -----------------------
# LƯU JSON
# -----------------------

def save_json():
    data = {
        "lop": [{"ma": lp.ma, "ten": lp.ten} for lp in lops],
        "sinhvien": [
            {"ma": sv.ma, "ten": sv.ten, "namsinh": sv.namsinh, "malop": sv.malop}
            for sv in sinhviens
        ]
    }
    with open(FILE_JSON, "w", encoding="utf-8") as f:
        json.dump(data, f, indent=4, ensure_ascii=False)
    print("Đã lưu JSON!")

def load_json():
    if not os.path.exists(FILE_JSON):
        return

    with open(FILE_JSON, "r", encoding="utf-8") as f:
        data = json.load(f)

        for lp in data["lop"]:
            lops.append(Lop(lp["ma"], lp["ten"]))
        for sv in data["sinhvien"]:
            sinhviens.append(SinhVien(sv["ma"], sv["ten"], sv["namsinh"], sv["malop"]))

def show_all():
    print("\n===== DANH SÁCH LỚP =====")
    for lp in lops:
        print(lp.ma, lp.ten)

    print("\n===== SINH VIÊN =====")
    for sv in sinhviens:
        print(sv.ma, sv.ten, sv.namsinh, "Lớp:", sv.malop)


# -----------------------
# CRUD
# -----------------------

def add_lop():
    ma = input("Mã lớp: ")
    ten = input("Tên lớp: ")
    lops.append(Lop(ma, ten))
    print("Thêm lớp thành công!")

def add_sv():
    ma = input("Mã SV: ")
    ten = input("Tên SV: ")
    ns = input("Năm sinh: ")
    malop = input("Mã lớp: ")
    sinhviens.append(SinhVien(ma, ten, ns, malop))
    print("Thêm sinh viên thành công!")

def search_sv():
    keyword = input("Nhập tên cần tìm: ")
    for sv in sinhviens:
        if keyword.lower() in sv.ten.lower():
            print(sv.ma, sv.ten, sv.namsinh, sv.malop)

# -----------------------
# MENU
# -----------------------

load_json()

while True:
    print("\n=== MENU QUẢN LÝ SINH VIÊN ===")
    print("1. Thêm lớp")
    print("2. Thêm sinh viên")
    print("3. Xem tất cả")
    print("4. Tìm kiếm sinh viên")
    print("5. Lưu JSON")
    print("0. Thoát")

    c = input("Chọn: ")

    if c == "1": add_lop()
    elif c == "2": add_sv()
    elif c == "3": show_all()
    elif c == "4": search_sv()
    elif c == "5": save_json()
    elif c == "0":
        save_json()
        break
```

```python
#Câu11:
from openpyxl import Workbook, load_workbook

FILE_NAME = "nhanvien.xlsx"

# =============================
# 1. HÀM LƯU NHÂN VIÊN VÀO EXCEL
# =============================
def luu_nhan_vien():
    ds_nv = [
        ["NV1", "An", 18],
        ["NV2", "Lành", 22],
        ["NV3", "Giải", 20],
        ["NV4", "Thoát", 19],
        ["NV5", "Hạnh", 25],
        ["NV6", "Phúc", 24]
    ]

    wb = Workbook()
    ws = wb.active
    ws.title = "NhanVien"

    ws.append(["STT", "Mã", "Tên", "Tuổi"])

    for i, nv in enumerate(ds_nv, start=1):
        ws.append([i] + nv)

    wb.save(FILE_NAME)
    print(">> Đã lưu danh sách vào", FILE_NAME)


# =============================
# 2. HÀM ĐỌC NHÂN VIÊN TỪ EXCEL
# =============================
def doc_nhan_vien():
    try:
        wb = load_workbook(FILE_NAME)
    except:
        print(">> Chưa có file Excel! Hãy lưu trước.")
        return

    ws = wb.active
    print("\n>> DANH SÁCH NHÂN VIÊN:")
    for row in ws.iter_rows(values_only=True):
        print(row)


# =============================
# 3. HÀM SẮP XẾP THEO TUỔI
# =============================
def sap_xep_nhan_vien():
    try:
        wb = load_workbook(FILE_NAME)
    except:
        print(">> Chưa có file Excel! Hãy lưu trước.")
        return

    ws = wb.active
    data = list(ws.iter_rows(values_only=True))

    header = data[0]
    rows = data[1:]

    rows_sorted = sorted(rows, key=lambda x: x[3])

    # Ghi ra file mới
    wb2 = Workbook()
    ws2 = wb2.active
    ws2.title = "SortedNhanVien"

    ws2.append(header)
    for r in rows_sorted:
        ws2.append(r)

    wb2.save("nhanvien_sorted.xlsx")
    print(">> Đã tạo file nhanvien_sorted.xlsx (sắp xếp theo tuổi)")


# =============================
# 4. MENU CHÍNH
# =============================
while True:
    print("\n===== QUẢN LÝ NHÂN VIÊN =====")
    print("1. Lưu nhân viên vào Excel")
    print("2. Đọc danh sách nhân viên")
    print("3. Sắp xếp theo tuổi tăng dần")
    print("0. Thoát")

    chon = input("Chọn chức năng: ")

    if chon == "1":
        luu_nhan_vien()
    elif chon == "2":
        doc_nhan_vien()
    elif chon == "3":
        sap_xep_nhan_vien()
    elif chon == "0":
        print("Thoát chương trình.")
        break
    else:
        print(">> Lựa chọn không hợp lệ!")
```

```python
#Câu12:
import csv
import random

# ============================
# 1. TẠO FILE CSV NGẪU NHIÊN
# ============================
def tao_file_csv(path):
    with open(path, "w", newline="") as f:
        writer = csv.writer(f, delimiter=';', quoting=csv.QUOTE_NONE)

        for _ in range(10):  # 10 dòng
            row = [str(random.randint(1, 100)) for _ in range(10)]  # 10 số ngẫu nhiên
            writer.writerow(row)

    print("Đã tạo file CSV thành công!")

# ============================
# 2. ĐỌC FILE CSV + TÍNH TỔNG
# ============================
def doc_file_va_tinh_tong(path):
    print("\n--- KẾT QUẢ TÍNH TỔNG MỖI DÒNG ---")
    with open(path, "r") as f:
        reader = csv.reader(f, delimiter=';')

        for i, row in enumerate(reader, start=1):
            numbers = list(map(int, row))     # chuyển về số nguyên
            tong = sum(numbers)
            print(f"Dòng {i}: {row}  --> Tổng = {tong}")

# ============================
#  CHẠY CHƯƠNG TRÌNH
# ============================
file_name = "data_random.csv"

tao_file_csv(file_name)
doc_file_va_tinh_tong(file_name)
```

```python
#Câu13:
from google.colab import files
uploaded = files.upload()
import os

print(os.listdir())

import xml.dom.minidom as minidom

def read_groups():
    groups = {}
    doc = minidom.parse("nhomthietbi.xml")
    nhoms = doc.getElementsByTagName("nhom")

    for nhom in nhoms:
        ma = nhom.getElementsByTagName("ma")[0].childNodes[0].data
        ten = nhom.getElementsByTagName("ten")[0].childNodes[0].data
        groups[ma] = ten
    return groups


def read_devices():
    devices = []
    doc = minidom.parse("ThietBi.xml")
    thietbis = doc.getElementsByTagName("thietbi")

    for tb in thietbis:
        manhom = tb.getAttribute("manhom")
        ma = tb.getElementsByTagName("ma")[0].childNodes[0].data
        ten = tb.getElementsByTagName("ten")[0].childNodes[0].data

        devices.append({
            "ma": ma,
            "ten": ten,
            "manhom": manhom
        })
    return devices


def print_devices_by_group():
    groups = read_groups()
    devices = read_devices()

    for manhom, tennhom in groups.items():
        print(f"\n=== {tennhom} ({manhom}) ===")
        for tb in devices:
            if tb["manhom"] == manhom:
                print(f" - {tb['ma']} : {tb['ten']}")

print_devices_by_group()
```
