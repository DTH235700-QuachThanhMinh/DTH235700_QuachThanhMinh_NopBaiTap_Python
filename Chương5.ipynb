```python
#câu4:
"""
- Trong Python, chuỗi (string) là kiểu dữ liệu quan trọng và được hỗ trợ bởi nhiều
hàm giúp việc xử lý văn bản trở nên dễ dàng hơn. Một số nhóm hàm thường được sử dụng gồm:
+ Python cung cấp các hàm thay đổi kiểu chữ như upper(), lower(), capitalize()…
giúp định dạng chuỗi theo yêu cầu, ví dụ chuyển thành chữ in hoa hoặc in thường.
Bên cạnh đó, các hàm kiểm tra (isalpha(), isdigit(), isspace()…) cho phép xác định chuỗi có phải là chữ cái, chữ số hay khoảng trắng.
Đây là các hàm thuận tiện khi cần kiểm tra dữ liệu đầu vào.
+ Python còn hỗ trợ các hàm tìm kiếm và thay thế như find() hay replace(), giúp xác định vị trí của một đoạn trong chuỗi hoặc thay thế nội dung.
Các hàm tách và nối chuỗi (split(), join()) đặc biệt hữu ích khi xử lý văn bản nhiều từ hoặc dữ liệu dạng câu.
Cuối cùng, nhóm hàm loại bỏ khoảng trắng (strip(), lstrip(), rstrip()) giúp làm sạch chuỗi trước khi xử lý.
"""
```

```python
#Câu5:
s = input("Nhập vào chuỗi: ")

# Khởi tạo biến đếm
hoa = thuong = so = dac_biet = khoang_trang = nguyen_am = phu_am = 0
nguyen_am_set = "aeiouAEIOU"

for ch in s:
    if ch.isupper():
        hoa += 1
    elif ch.islower():
        thuong += 1

    if ch.isdigit():
        so += 1
    elif ch.isspace():
        khoang_trang += 1
    elif not ch.isalnum() and not ch.isspace():
        dac_biet += 1

    if ch.isalpha():  # nếu là chữ
        if ch in nguyen_am_set:
            nguyen_am += 1
        else:
            phu_am += 1

# Xuất kết quả
print("Số chữ IN HOA:", hoa)
print("Số chữ in thường:", thuong)
print("Số chữ là chữ số:", so)
print("Số ký tự đặc biệt:", dac_biet)
print("Số ký tự khoảng trắng:", khoang_trang)
print("Số nguyên âm:", nguyen_am)
print("Số phụ âm:", phu_am)
```

```python
#Câu6:
def NegativeNumberInStrings(s):
    numbers = []
    i = 0
    while i < len(s):
        if s[i] == '-' and i + 1 < len(s) and s[i+1].isdigit():
            j = i + 1
            while j < len(s) and s[j].isdigit():
                j += 1
            numbers.append(int(s[i:j]))  # ép thành số nguyên âm
            i = j  # nhảy tới sau số
        else:
            i += 1
    return numbers

# Test
chuoi = "abc-5xyz-12k9l--p"
print(NegativeNumberInStrings(chuoi))  # [ -5, -12 ]
```

```python
#Câu8:
import os

# Hàm lấy tên file kèm đuôi
def lay_ten_file(path):
    return os.path.basename(path)   # ví dụ: muabui.mp3

# Hàm lấy tên file không có đuôi
def lay_ten_bai_hat(path):
    return os.path.splitext(os.path.basename(path))[0]  # ví dụ: muabui


# --- Test thử ---
duong_dan = r"d:\music\muabui.mp3"
print("Tên file:", lay_ten_file(duong_dan))      # muabui.mp3
print("Tên bài hát:", lay_ten_bai_hat(duong_dan)) # muabui
```
