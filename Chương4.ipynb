```python
#Câu 1: Viết hàm để tính diện tích tam giác.
from math import sqrt
print("Chương trình tính diện tích Tam Giác: ")
a = float(input("Nhập cạnh a > 0: "))
b = float(input("Nhập cạnh b > 0: "))
c = float(input("Nhập cạnh c > 0: "))
if (a <= 0 or b <= 0 or c <= 0) or (a + b) <= c or (a + c) <= b or (b + c) <= a:
  print("Tam giác không hợp lệ")
else:
  cv = a + b + c
  p = cv/2
  dt=sqrt(p * (p - a) * (p - b) * (p - c))
  print("Diện tích = ",dt)
```

```python
#Câu 4: Viết Hàm tính ROI.
def ROI(dt,cp):
    return (dt - cp)/cp
def GoiYDauTu(roi):
    if roi >= 0.75:
      return "Nên đầu tư"
    else:
      return "Không nên đầu tư"
print("Chương trình tính ROI: ")
dt = int(input("Nhập doanh thu: "))
cp = int(input("Nhập chi phí: "))
roi = ROI(dt,cp)
print("Tỉ Lệ ROI = ",roi)
print("==>",GoiYDauTu(roi))
```

```python
#Câu 9: Viết chương trình tính căn bậc 2 lồng nhau.
import math

def nested_root_iter(n):
    """Tính căn lồng nhau bằng cách lập (từ trong ra ngoài)."""
    if n <= 0:
        return 0.0
    val = math.sqrt(2.0)  # n = 1
    for _ in range(2, n+1):
        val = math.sqrt(2.0 + val)
    return val

def nested_root_formula(n):
    """Công thức lượng giác: 2 * cos(pi / 2^(n+1))"""
    return 2.0 * math.cos(math.pi / (2 ** (n + 1)))

def main():
    n = int(input("Nhập n (số dấu căn): ").strip())
    if n <= 0:
        print("n phải là số nguyên dương.")
        return
    it = nested_root_iter(n)
    fm = nested_root_formula(n)
    print(f"Kết quả (lặp): {it}")
    print(f"Kết quả (công thức): {fm}")
    print(f"Hiệu tuyệt đối: {abs(it - fm)}")

if __name__ == "__main__":
    main()
```

```python
#Câu 13: Hàm kiểm tra số hoàn thiện, số thịnh vượng.
# Bước 1: Viết hàm tính tổng ước số
def tong_uoc_so(n):
    if n <= 0:
        return 0
    tong = 0

    for i in range(1, int(n**0.5) + 1):
        if n % i == 0:

            tong += i
            if i != n // i and i != 1:
                 tong += n // i
            elif i == n // i and i != 1:
                 pass
            elif i == 1 and n != 1:
                 pass
            elif n == 1:
                 tong = 0
                 break
    if n > 1:
      return tong - n
    else:
      return tong

# Bước 2: Viết hàm kiểm tra số hoàn thiện
def so_hoan_thien(n):
    if n <= 0:
        return False
    return tong_uoc_so(n) == n


# Bước 3: Viết hàm kiểm tra số thịnh vượng
def so_thinh_vuong(n):
    if n <= 0:
        return False
    return tong_uoc_so(n) > n

# Bước 4: Kiểm tra và hiển thị kết quả
def main_perfect_abundant():
    try:
        nhap_so = input("Nhập một số nguyên dương để kiểm tra: ").strip()
        so_can_kiem_tra = int(nhap_so) # Chuyển đổi sang số nguyên

        # Kiểm tra điều kiện số nguyên dương
        if so_can_kiem_tra <= 0:
            print("Yêu cầu nhập một số nguyên dương.")
            return # Dừng chương trình nếu không hợp lệ

        # Sử dụng các hàm đã tạo để kiểm tra
        if so_hoan_thien(so_can_kiem_tra):
            print(f"{so_can_kiem_tra} là số hoàn thiện.")
        elif so_thinh_vuong(so_can_kiem_tra):
            print(f"{so_can_kiem_tra} là số thịnh vượng.")

    except ValueError:
        # Xử lý trường hợp người dùng nhập không phải số nguyên
        print("Đầu vào không hợp lệ. Vui lòng nhập một số nguyên.")

# Chạy hàm main khi script được thực thi trực tiếp
if __name__ == "__main__":
    main_perfect_abundant()
```
